# Résultats de génération de musique
Ce fichier présente nos résultats de musiques générées avec différents modèles.

Sommaire:
* Aléatoire
* VAE
* LSTM
* GAN

```python
import music21 as m21

def listenTo(path):
    mf = m21.midi.MidiFile()
    mf.open(path)
    mf.read()
    mf.close()
    song = m21.midi.translate.midiFileToStream(mf)
    song.show()
    song.show("midi")
```

# Aléatoire

```python
listenTo("ALEA/random.mid")
```

```python
listenTo("ALEA/random_with_chords.mid")
```
