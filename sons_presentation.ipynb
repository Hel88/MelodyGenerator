```python
import music21 as m21
```

```python
def showSong(path):
    mf = m21.midi.MidiFile()
    mf.open(path)
    mf.read()
    mf.close()
    song = m21.midi.translate.midiFileToStream(mf)
    song.show()
    song.show("midi")
```

```python
showSong("generated_melodies/RNN-LSTM/deutsch/0604_173326_8.mid")
```
